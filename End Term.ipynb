```python
def Histogram(data):
    non_floats = []
    for col in data:
        if data[col].dtypes != "int64":
            if data[col].dtypes != "float64":
                non_floats.append(col)
    data = data.drop(columns=non_floats)
    for col in data:
        if data[col].dtypes == "float64" or "int64":
            data.hist(column = col, grid = False)
            plt.savefig('plothist'+str(col)+'.png')
            plt.show()
            plt.close()
```

```python
Histogram(data)
```

```python
def Graphs(data):
    non_floats = []
    for col in data:
        if data[col].dtypes != "int64":
            if data[col].dtypes != "float64":
                non_floats.append(col)
    df = data.drop(columns=non_floats)
    for col in df:
        if df[col].dtypes == "float64" or "int64":
            plt.rcParams["figure.figsize"] = (8, 6)
            plt.rcParams["xtick.labelsize"] = 7
            df.hist(column = col, grid = False)
            plt.savefig('plothist'+str(col)+'.png')
            plt.show()
            plt.close()
    
    df = data.drop(columns=non_floats)
    for col in df:      
        if df[col].dtypes == "float64" or "int64":
            plt.figure(figsize=(8,6))
            sns.boxplot(y = df[col])
            plt.savefig('plotbox'+str(col)+'.png')
            plt.show()
            plt.close()
            
    categ=data.dtypes[data.dtypes=='object'].index
    for col in categ:
        fig = plt.figure(figsize=(30,5))
        plot = sns.countplot(data[col])
        plot.set_xticklabels(plot.get_xticklabels(),rotation=90, ha="right")
        plt.savefig('plotcateg'+str(col)+'.png')
        plt.show()
        plt.close()

            
            
```

```python
Graphs(data)
```

```python
#This function gives the basic statistics of each of the variables in the data set 

def Simple_Stat(data, var = list(range(len(data.columns)))):
    if var == []:
        print(Graphs(data))
    data = data.iloc[:, var]
    non_floats = []
    for col in data:
        if data[col].dtypes != "int64":
            if data[col].dtypes != "float64":
                non_floats.append(col)
    df = data.drop(columns=non_floats)
    for col in df:
        if df[col].dtypes == "float64" or "int64":
            plt.rcParams["figure.figsize"] = (8, 6)
            plt.rcParams["xtick.labelsize"] = 7
            df.hist(column = col, grid = False)
            print(df[col].describe())
            plt.savefig('plothist'+str(col)+'.png')
            plt.show()
            plt.close()
            
    
    df = data.drop(columns=non_floats)
    for col in df:      
        if df[col].dtypes == "float64" or "int64":
            plt.figure(figsize=(8,6))
            sns.boxplot(y = df[col])
            print(df[col].describe())
            plt.savefig('plotbox'+str(col)+'.png')
            plt.show()
            plt.close()
           
            
    categ=data.dtypes[data.dtypes=='object'].index
    for col in categ:
        fig = plt.figure(figsize=(30,5))
        plot = sns.countplot(data[col])
        plot.set_xticklabels(plot.get_xticklabels(),rotation=90, ha="right")
        tab = pd.crosstab(data[col], columns='Frequency')
        print(tab)
        plt.savefig('plotcateg'+str(col)+'.png')
        plt.show()
        plt.close()
    
            
```
